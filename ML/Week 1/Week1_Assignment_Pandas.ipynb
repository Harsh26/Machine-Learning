```python
null_vals = 0
tar_col = None

for col in df:
    print(col, df[col].isnull().sum(), sep=':')
    
    if df[col].isnull().sum() > null_vals:
        null_vals = df[col].isnull().sum()
        tar_col = col

print(tar_col)

if tar_col is not None:
    df = df.drop(tar_col, axis=1)

df
```

```python
print(df.describe(include='all').dtypes)
```

```python
males = 0
females = 0

for row in df['Sex']:
        if row == 'male':
            males += 1
        else:
            females += 1
print('female / male = ', females, '/', males, sep = ' ')
print('Sex Ratio:', females / males, sep=" ")
```

```python
# Sex ratio by value_count()

mf = df['Sex'].value_counts()

# print(mf)
# print(list(mf))

mfl = list(mf)

xy = mfl[0]
xx = mfl[1]

print('Sex Ratio:', xx / xy, sep=" ")
```

```python
survived = df['Survived'].sum()
print('No. of people survived = ', survived)
```

```python
maxage = (df['Age'].max())
minage = (df['Age'].min())

print(maxage, minage, sep = " ")
```

```python
# Efficent method to print rows

maxageid = df['Age'].idxmax()
minageid = df['Age'].idxmin()

maxagerow = df.loc[maxageid]
minagerow = df.loc[minageid]

print('Max Age row: \n', maxagerow)
print('Min Age row: \n', minagerow)
```

```python
count2 = 0

for c2 in df['Pclass']:
    if c2 == 2:
        count2 += 1

print(count2)
```

```python
sf = 0
sm = 0

for sex, surv in zip(df['Sex'], df['Survived']):
    if surv == 1 and sex == 'female':
            sf += 1
    
    if surv == 1 and sex == 'male':
            sm += 1

print('Females Survived = ', sf)
print('Males Survived = ', sm)
```

```python
p = 0

for surv, fare in zip(df['Survived'], df['Fare']):
    if surv == 1 and fare < 200:
        p += 1

print('Persons survived with less fare = ', p)
```
